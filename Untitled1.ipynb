```python
### PRÁCTICA DE CREACIÓN DE FICHEROS ###
### SE VA A CREAR UN FICHERO ALMACENANDO LOS VALORES DEL ÁREA DE UNA 
### FUNCIÓN, OBTENIDOS A TRAVÉS DE LA SUMA DE RIEMANN , ESPECÍFICAMENTE,
### LA SUMA INFERIOR DE LOS N RECTÁNGULOS DEL ÁREA DE LA REGIÓN EN EL INTERVALO
### [a, b] DE LA FUNCIÓN DADA.
```

El área a calcular, será para la región formada entre la curva $f(x) = x^{1/2} + 1$, y las rectas $x=2$ y $x=5$:

Área debajo de una curva

```python
## creando fichero 
f = open("area.txt", "w") 

## el programa va a calcular el área, iterando entre 1 y 5000 rectángulos N,
## que dividirán a la región y luego hará una sumatoria del área de cada rectágulo
## para obtener el área total de la región.

for N in range(1, 5001):
  area = 0
  ancho = 3 / N  # calcula el ancho de rectángulo 
   
  for rec in range(N): # recorreremos cada rectángulo 
    x = 2 + (rec * (3/N)) # calculando la coordenada x del rectágulo actual

    alto = x ** 0.5 + 1   # calculamos el alto del rectángulo para el x anterior

    area = area + ancho * alto # calculando el área de cada rectángulo y sumádolo al area total

  print(f"{N} {area:.4f}", file = f)  #Imprimiendo al fichero una nueva línea con dos valores:
                                       # el número de rectángulos y el valor del área con 4 decimales.

f.close()  #cierre del fichero                                
```

```python
## vemos en el fichero que el área proximada de la región para 5000 rectángulos es de 8.5677
```
